```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
train_df = pd.read_csv('fraudTrain.csv')
test_df = pd.read_csv('fraudTest.csv')
```

```python
drop_cols = ['Unnamed: 0','trans_date_trans_time','cc_num','first','last','street','city','state','zip','dob','trans_num']
```

```python
train_df.drop(columns=drop_cols, inplace=True)
test_df.drop(columns=drop_cols, inplace=True)
```

```python
categorical_cols = train_df.select_dtypes(include='object').columns
```

```python
for col in categorical_cols:
    le = LabelEncoder()

    combined = pd.concat([train_df[col], test_df[col]])

    le.fit(combined)

    train_df[col] = le.transform(train_df[col])
    test_df[col] = le.transform(test_df[col])
```

```python
X_train = train_df.drop('is_fraud', axis=1)
y_train = train_df['is_fraud']

X_test = test_df.drop('is_fraud', axis=1)
y_test = test_df['is_fraud']
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
clf1=LogisticRegression(max_iter=1000)
clf1.fit(X_train,y_train)
pred1=clf1.predict(X_test)
print("Accuracy Score of Logistic Regression is:",accuracy_score(y_test,pred1))
```

```python
confusion_matrix(y_test,pred1)
```

```python
print(classification_report(y_test, pred1))
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
clf2=DecisionTreeClassifier(random_state=42)
clf2.fit(X_train,y_train)
pred2=clf2.predict(X_test)
print("Accuracy Score of Decision Tree is:",accuracy_score(y_test,pred2))
```

```python
confusion_matrix(y_test, pred2)
```

```python
print(classification_report(y_test,pred2))
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
clf3=RandomForestClassifier(n_estimators=100,random_state=42)
clf3.fit(X_train,y_train)
pred3=clf3.predict(X_test)
print("Accuracy Score of Random Forest is:",accuracy_score(y_test,pred3))
```

```python
confusion_matrix(y_test, pred3)
```

```python
print(classification_report(y_test,pred2))
```
